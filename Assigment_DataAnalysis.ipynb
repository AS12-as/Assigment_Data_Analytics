```python
Mean_Rating= Order_Details['Rating'].mean()
```

```python
Mean_Rating
```

```python
# Replace null values in 'Rating' with the mean
Order_Details['Rating'] = Order_Details['Rating'].fillna(round(Mean_Rating))
Order_Details
```

```python
#As due to Cancellation Of the order User have not give the rating but analyse a pattern of the Data
```

```python
# by using user_id as a key to merge data
merged_data = pd.merge(Order_Details, Cooking_Session, on="Session ID", how="inner")
merged_data
```

```python
merge_data_df = merged_data.drop(["User ID_y","Meal Type_y","Dish Name_y"], axis=1)
```

```python
merge_data_df
```

```python
rename_data_df= merge_data_df.rename(columns={"User ID_x": "User ID", "Meal Type_x": "Meal Type", "Dish Name_x":"Dish Name"})
rename_data_df
```

```python
data_df= rename_data_df
```

```python
rename_data_df
```

```python
data_df
```

```python
merge_data= pd.merge(data_df, user_details, on="User ID", how="inner")
```

merge_data

```python
merge_data
```

```python
#Relationship Between Cooking Sessions, Orders, Dishes and Amount
cooking_order_analysis = merge_data.groupby('User ID').agg({
    'Order ID': 'count',
    'Session ID': 'count',
    'Dish Name' : 'count',
    'Amount (USD)' : 'sum'
}).rename(columns={'Order ID': 'Total Orders', 'Session ID': 'Total Sessions','Dish Name': 'Total Dishes',  'Amount (USD)': 'Total Amount'})
print(cooking_order_analysis)
```

```python
#Identify Most Popular Dishes
popular_dishes = merge_data['Dish Name'].value_counts()
print(popular_dishes)
```

```python
#Demographic Insights: Analyze order trends by Favourite Meal, location, age and username total orders as per differenct location 
demographic_analysis = user_details.groupby(['Favorite Meal', 'Location', 'User Name', 'Age' ])['Total Orders'].sum()
print(demographic_analysis)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

popular_dishes.head(10).plot(kind='bar', title='Most Popular dishes')
plt.show()

```

```python
#Demographic Influence on Orders shows that people favourite meal to order at dinner
favorite_meal_orders = user_details.groupby("Favorite Meal")["Total Orders"].sum()

# Plot the bar chart
plt.figure(figsize=(8, 5))
favorite_meal_orders.plot(kind="bar", color="cornflowerblue", edgecolor="black")
plt.title("Favorite Meal vs Total Orders", fontsize=14)
plt.xlabel("Favorite Meal", fontsize=12)
plt.ylabel("Total Orders", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#Demographic Influence on Orders
sns.boxplot(data=user_details, x='Favorite Meal', y='Total Orders')
plt.title('Favorite Meal vs Total Orders')
plt.show()
```

```python
#session duration and rating
import matplotlib.pyplot as plt

# Assuming merged_data contains the relevant data
plt.figure(figsize=(8, 6))

# Bubble chart with bubble size representing 'Amount (USD)'
plt.scatter(
    merged_data['Duration (mins)'], 
    merged_data['Session Rating'], 
    s=merged_data['Amount (USD)'] * 10,  # Scale bubble size
    alpha=0.6, 
    color="cornflowerblue", 
    edgecolor="black"
)

# Customize the plot
plt.title("Session Duration vs Ratings ", fontsize=14)
plt.xlabel("Duration (mins)", fontsize=12)
plt.ylabel("Session Rating", fontsize=12)
plt.grid(alpha=0.5)

# Show the chart
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt


#  Group by 'Favorite Meal' and 'Location' and sum 'Total Orders'
demographic_analysis = user_details.groupby(['Favorite Meal', 'Location'])['Total Orders'].sum().unstack()

# Plot a stacked bar chart
demographic_analysis.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')

# Step 3: Customize the chart
plt.title("Demographic Analysis: Total Orders by Favorite Meal and Location", fontsize=14)
plt.xlabel("Favorite Meal", fontsize=12)
plt.ylabel("Total Orders", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Location", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
